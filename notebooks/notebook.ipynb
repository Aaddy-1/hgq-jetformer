```python
import os
os.environ["KERAS_BACKEND"] = "jax"

from hgq.config import LayerConfigScope, QuantizerConfigScope
import keras
import numpy as np
import sys

# Adds the parent directory to the python path
sys.path.append(os.path.abspath(".."))

from src.models.jetformer import HGQJetFormer
```

```python

# 1. Define dummy data (1 jet, 30 particles, 16 features)
dummy_input = np.random.rand(1, 30, 16).astype("float32")

# 2. Instantiate within Scopes
with (
    QuantizerConfigScope(place='all', default_q_type='kbi', overflow_mode='SAT_SYM'),
    LayerConfigScope(enable_ebops=True, beta0=1e-5)
):
    model = HGQJetFormer(in_dim=16, embed_dim=128)
    
    # Run the pass
    output = model(dummy_input, training=False)

print(f"Success! Output shape: {output.shape}") # Should be (1, 5)
```
